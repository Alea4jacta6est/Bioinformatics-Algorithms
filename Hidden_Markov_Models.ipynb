```python
def Viterbi_FB(seq, transition, initials, generation):
    
    '''
    Gets most probable sequence of hidden states and estimates most probable hidden
    state at each time step.
    '''
    
    def forward_pass(MP_states, i):
        pass
    
    def backward_pass(MP_states, i):
        pass
    
    n=len(seq)
    m=len(initials)
    
    MP_seq=[]
    MP_states=np.zeros([m, n])
    
    MAX_prob=np.zeros([m, n])
    MAX_prob[:, 0]=[initials[j]*generation.loc[j+1, seq[0]] for j in range(m)]
    MAX_state=np.zeros([m, n], dtype='int')
    
    Forward_prob=np.zeros([m, n])
    Forward_prob[:, 0]=[initials[j]*generation.loc[j+1, seq[0]] for j in range(m)]
    Backward_prob=np.zeros([m, n])
    Backward_prob[:, n-1]=[1]*m
    
    for i in range(1, n):
        MAX_prob[:, i]=[max([MAX_prob[k, i-1]*transition[k, j]*generation.loc[j+1, seq[i]] \
                             for k in range(m)]) for j in range(m)]
        MAX_state[:, i]=[np.argmax([MAX_prob[k, i-1]*transition[k, j]*generation.loc[j+1, seq[i]] \
                                    for k in range(m)])+1 for j in range(m)]
        
        Forward_prob[:, i]=[sum([Forward_prob[k, i-1]*transition[k, j]*generation.loc[j+1, seq[i]] \
                             for k in range(m)]) for j in range(m)]
        Backward_prob[:, (n-1)-i]=[sum([Backward_prob[k, n-i]*transition[j, k]*generation.loc[k+1, seq[n-i]] \
                             for k in range(m)]) for j in range(m)]
        
    CT=sum(Forward_prob[:, n-1])
    for t in range(n):
        MP_states[:, t]=[(Forward_prob[j, t]*Backward_prob[j, t])/CT\
                         for j in range(m)]
        
    MP_seq.append(np.argmax(MAX_prob[:, n-1])+1)
    
    for i in range(n-2, 0, -1):
        MP_seq.append(MAX_state[MP_seq[-1]-1, i])
        
    return ''.join([str(x) for x in MP_seq[::-1]]), pd.DataFrame(MP_states, index=generation.index, \
                                                                 columns=['t_{}'.format(x) for x in range(1, n+1)])
```
